```python
#Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#Loading company and rounds2 data into dataframe
company_df = pd.read_csv('companies.txt', sep='\t', encoding='ISO-8859-1')
rounds2_df = pd.read_csv('rounds2.csv', sep=',', encoding='ISO-8859-1')

print(company_df.head(5))
print(company_df.info())
print(rounds2_df.head(5))
print(rounds2_df.info())

#Converting permalink and country_permalink column to upper case to make the format consistent
company_df['permalink']=company_df['permalink'].apply(lambda x:x.upper())
rounds2_df['company_permalink']=rounds2_df['company_permalink'] .apply(lambda x:x.upper())

print(company_df.head(5))
print(rounds2_df.head(5))
```

```python
#Checkpoint 1

#Get unique companies in company and rounds2 data frames

print('\n No of unique Companies in company_df \n', company_df['permalink'].describe())
print('\n No of unique companies in rounds2_df \n', rounds2_df['company_permalink'].describe())

#Get list of companies which are there in rounds2 but not in company dataframe
rounds2_df[~rounds2_df['company_permalink'].isin(company_df['permalink'])]

#Rename 'company-permalink' column to 'permalink' to facilitate merge
rounds2_df = rounds2_df.rename(columns={'company_permalink': 'permalink'})
print(rounds2_df.head(5))

#Merging company_df with rounds2_df on permalink column
master_frame = pd.merge(rounds2_df, company_df, how='inner', on='permalink')
master_frame.head(5)
```

```python
#Get number of observations in master dataframe
print(master_frame.shape)
print(master_frame.info())

#Get percentage of null values in master dataframe columns 
print(master_frame.isnull().sum())
print('\n Null % \n',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))

#Drop the columns not necessary for this analysis and re check the null percentage
master_frame=master_frame.drop('funding_round_code', axis=1)
master_frame=master_frame.drop('founded_at', axis=1)
master_frame=master_frame.drop('homepage_url', axis=1)
master_frame=master_frame.drop('state_code', axis=1)
master_frame=master_frame.drop('region', axis=1)
master_frame=master_frame.drop('city', axis=1)

print('\n Null % after column drop \n',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))

#Check new dimenssions of master frame
print(master_frame.shape)
print(master_frame.info())
```

```python
#Get number of rows with null values in more than three columns 
print('\n No of null values in more than 3 columns \n ',master_frame[master_frame.isnull().sum(axis=1)>=3].isnull().sum())

#Drop the rows since the data is missing in all the rows with the same columns and check null percentage
master_frame = master_frame[master_frame.isnull().sum(axis=1)<3]

print('\n Null % post rows drop with null values >= 3 \n ',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))

# Check mean, standard deviation to check feasibility of imputing missing values 
print('\n Observations on master_frame \n', master_frame.raised_amount_usd.describe())

#Since the standard deviation is huge drop the rows and re check the null percentage
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
print('\n Null % after droping null rows of raised_amount_usd \n',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))
```

```python
#Check the number of rows with null values in category_list and country_code column

print(master_frame.category_list.isnull().sum())
print(master_frame.country_code.isnull().sum())

#Remove the rows with null values in country_code column and re check the null percentage

master_frame=master_frame[~(pd.isnull(master_frame['country_code']))]
print('\n Null % after dropping null rows in country_code \n',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))

#Check the number of rows with null values in category_list and remove them
print(master_frame.category_list.isnull().sum())

#Remove the rows with null values in category_list and recheck the null percentage
master_frame=master_frame[~pd.isnull(master_frame['category_list'])]

print('\n Null % after dropping category_list null rows \n',round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2))

#Check percentage of data retained
print('% of data retained \n',round(100*len(master_frame.index)/114942,2))
```

```python
#Checkpoint 2

#Changing the floating point display format for better readability
pd.set_option('display.float_format', lambda x: '%.3f' % x)

#Get the details of funding_round_type
print(master_frame.groupby('funding_round_type').mean())

#Get the no of investments based on funding_round_type
master_frame.groupby('funding_round_type')['permalink'].count()

```

```python
#Venture type is the most preferred as the mean falls in the range of 5 to 15 million and no of investments is also high
venture_df = master_frame.loc[(master_frame['funding_round_type'] =='venture')]
venture_df.head()
```

```python
#Checkpoint 3

#Group the data based on country code
df_by_country=venture_df.groupby('country_code')

#Perform aggregation on grouped values and store top 9 records in a dataframe after sorting based on funding amount
top9=df_by_country.raised_amount_usd.sum().sort_values(ascending=False).head(9)
top9.head(9)
```

```python
#Create a new data frame with top 3 English speaking countries with highest investment
top3_country_df = venture_df.loc[(venture_df.country_code == 'USA') | (venture_df.country_code == 'GBR') | (venture_df.country_code == 'IND')]
top3_country_df.head(10)
```

```python
#Split the category list based on delimitter and add it to a new dataframe
primary_cat_series = top3_country_df['category_list'].apply(lambda x: x.split('|')[0])
primary_cat_series.head(10)

```

```python
#Concat the two data frames into a new dataframe. Rename the new column to primary_category
concat_cat_df=pd.concat([top3_country_df, primary_cat_series.rename('primary_sector')], axis=1)

concat_cat_df.head(20)
```

```python
#Read mapping csv file
mapping_df = pd.read_csv('mapping.csv')

print(mapping_df.info())

mapping_df.head()
```

```python
# store the value and id variables in two separate arrays

# store the value variable in one Series
value_vars = mapping_df.columns[1:]

# take the setdiff() to get the rest of the variables
id_vars = np.setdiff1d(mapping_df.columns, value_vars)

print(value_vars, "\n")
print(id_vars)
```

```python
#Converting the wide dataframe format to long dataframe format

mapping_long_df= pd.melt(mapping_df, id_vars=list(id_vars), value_vars=list(value_vars))

mapping_long_df.head(20)
```

```python
#Take only the values with data as 1
mapping_long_df=mapping_long_df[mapping_long_df['value']==1]

#Rename the columns to match the main dataframe
mapping_long_df = mapping_long_df.rename(columns={'variable': 'main_sector', 'category_list':'primary_sector', 'value':'counts'})
print(mapping_long_df.head(100))
```

```python
#Checkpoint 4

#Merge the two data frames on primary_category column
primary_sector_merge = pd.merge(concat_cat_df, mapping_long_df, how='inner', on='primary_sector')


primary_sector_merge.head(10)
```

```python
#Checkpoint 5

#Filtering the values based on top three countries and interested investment value between 5 and 15 million

us_df = primary_sector_merge.loc[(primary_sector_merge.raised_amount_usd >= 5000000) & (primary_sector_merge.raised_amount_usd <= 15000000) & (primary_sector_merge.country_code=='USA')]
br_df = primary_sector_merge.loc[(primary_sector_merge.raised_amount_usd >= 5000000) & (primary_sector_merge.raised_amount_usd <= 15000000) & (primary_sector_merge.country_code=='GBR')]
in_df = primary_sector_merge.loc[(primary_sector_merge.raised_amount_usd >= 5000000) & (primary_sector_merge.raised_amount_usd <= 15000000) & (primary_sector_merge.country_code=='IND')]

print(us_df.head())

# store the value and id variables in two separate arrays

# store the value variable in one Series
pivot_value_vars = pd.Series(['raised_amount_usd','counts'])

# take the setdiff() to get the rest of the columns
pivot_id_vars = np.setdiff1d(primary_sector_merge.columns[0:-2], pivot_value_vars)

print(pivot_value_vars, "\n")
print(pivot_id_vars)

#Convert to pivot table to view the data in wide format

D1 = pd.pivot_table(us_df, index=list(pivot_id_vars),columns=['main_sector'], values=list(pivot_value_vars), aggfunc=np.sum)

D2 =  pd.pivot_table(br_df, index=list(pivot_id_vars),columns=['main_sector'], values=list(pivot_value_vars), aggfunc=np.sum)

D3 =  pd.pivot_table(in_df, index=list(pivot_id_vars),columns=['main_sector'], values=list(pivot_value_vars), aggfunc=np.sum)


D1.head()



```

```python
#Analysis for US

#Total number of investments
print(pd.Series(D1['counts'].describe().sum()).sum())

#Total amount of investment
print(pd.Series(D1['raised_amount_usd'].describe().sum()).sum())

#Top sector, second sector, third sector based on no of investments
print(D1['counts'].describe().loc[['count']])

#Company receiving highest investment for top sector ('Others')
print(us_df[us_df['main_sector']=='Others'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())

#Company receiving highest investment for second best sector ('Cleantech / Semiconductors')
print(us_df[us_df['main_sector']=='Cleantech / Semiconductors'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())
```

```python
#Analysis for GBR

#Total number of investments
print(pd.Series(D2['counts'].describe().sum()).sum())

#Total amount of investment
print(pd.Series(D2['raised_amount_usd'].describe().sum()).sum())

#Top sector, second sector, third sector based on no of investments
print(D2['counts'].describe().loc[['count']])

#Company receiving highest investment for top sector ('Others')
print(br_df[br_df['main_sector']=='Others'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())

#Company receiving highest investment for second best sector ('Cleantech / Semiconductors')
print(br_df[br_df['main_sector']=='Cleantech / Semiconductors'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())
```

```python
#Analysis for India

#Total number of investments
print(pd.Series(D3['counts'].describe().sum()).sum())

#Total amount of investment
print(pd.Series(D3['raised_amount_usd'].describe().sum()).sum())

#Top sector, second sector, third sector based on no of investments
print(D3['counts'].describe().loc[['count']])

#Company receiving highest investment for top sector ('Others')
print(in_df[in_df['main_sector']=='Others'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())

#Company receiving highest investment for second best sector (News, Search and Messaging)
print(in_df[in_df['main_sector']=='News, Search and Messaging'].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False).head())
```

```python
#Checkpoint 6

#Get data pertaining to venture, seed and private equity funding types
filtered_df = master_frame.loc[(master_frame['funding_round_type']=='venture') | (master_frame['funding_round_type']=='seed') | (master_frame['funding_round_type']=='private_equity')]

print(filtered_df.groupby('funding_round_type').mean())

# set seaborn theme
sns.set(style="white")

#Box and count plot to show average investments and number of investments


plt.subplots(figsize=(20,10))
# subplot 1: Mean
plt.subplot(1, 3, 1)
axis_bar = sns.barplot(x='funding_round_type', y='raised_amount_usd', data=filtered_df)
axis_bar.set(xlabel='Investment Type', ylabel='Average Investment Amount (USD) in Millions')
plt.title('Average investment amount across investment types')

# subplot 2: No of investments
plt.subplot(1, 3, 3)
axis_count = sns.countplot(x='funding_round_type' , data=filtered_df)
axis_count.set(xlabel='Investment Type', ylabel='No of Investments')
plt.title('No of investments')
plt.show()
```

```python
#Get the data frame consisting of top 9 countries investment
top9_df = pd.DataFrame({'country_code':top9.index, 'raised_amount_usd':top9.values})

plt.subplots(figsize=(10,8))

#Bar plot for total no of investments by country code
axis = sns.barplot(x='country_code', y='raised_amount_usd', data=top9_df)
plt.yscale('log')
axis.set(xlabel='Country Code', ylabel='Investment Amount Millions of USD')
plt.title('Amount of Investments by country')
plt.show()
```

```python
#Get all the sectors with investment range between 5 to 15 million
sector_df = primary_sector_merge.loc[(primary_sector_merge.raised_amount_usd >= 5000000) & (primary_sector_merge.raised_amount_usd <= 15000000)]

#Get the investments in the main sectors by the country code
top3_df = sector_df['main_sector'].groupby(sector_df['country_code']).value_counts()

#Get the top 3 sectors in top 3 countries in a dataframe
top3_filtered = pd.DataFrame({'count' : top3_df.groupby( 'country_code').head(3)}).reset_index()

plt.subplots(figsize=(10,8))
#Plot to show the no of investments in top 3 countries in top 3 sectors
axis = sns.barplot(x='country_code', y='count', hue='main_sector', data=top3_filtered)
plt.yscale('log')
axis.set(xlabel='Country Code', ylabel='No of investments')
plt.title('No of investments in top 3 countries in top 3 sectors')
plt.show()
```
