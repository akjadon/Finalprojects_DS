#### Import Libraries and read the csv file


```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Read CSV to dataframe
loan_df = pd.read_csv('loan.csv', sep=',', encoding='ISO-8859-1')

loan_df.head(50)
```


Checking for the column which represents unique data across rows. We get 39717 entries in the dataframe and 'id' column has 39717 unique entries.


```python
#Observations in loan dataframe. Get unique rows
loan_df.info()
print(loan_df['id'].nunique())
```

```python
#Check the number of records with null values
loan_df.isnull().sum()
```

Get the null percentage in the dataframe

```python
#Get the percentage of rows with null values
round(100*(loan_df.isnull().sum()/len(loan_df.index)), 2)
```

Removing the columns which are completely empty and re check the null percentage

```python
#Get the columns whose null values sum is not equal to 39717. This removes the columns with 100% null percentage
loan_drop_null_df = loan_df.loc[:, loan_df.isnull().sum()!=39717]

loan_drop_null_df.isnull().sum()
```

```python
#Check the null percentage
round(100*(loan_drop_null_df.isnull().sum()/len(loan_drop_null_df.index)), 2)
```

```python
loan_drop_null_df.corr().to_csv('correlation.csv')
```

#### Drop the un necessary columns

We are only interested about the columns which are having impact on the 'Charged Off' loan status

loan_amnt,funded_amnt, funded_amnt_inv, installment, total_pymnt, total_pymnt_inv, total_rec_prncp, total_rec_int are highly correlated. Analysis on any of these columns gives same results.
loan_amnt and int_rate are retained and rest of the columns can be removed.

Similarly mths_since_last_record,mths_since_last_delinq, pub_rec and pub_rec_bankruptcies are highly correlated. Since pub_rec_bankruptcies is having more null values only pub_rec is retained and other columns are removed.

revol_bal, revol_util represent same. revol_bal is removed.

member_id and id both represent unique rows. member_id is removed.

next_pymnt_d,  - It has no impact on the current loan status to be charged off. So this can be removed.

out_prncp_inv, total_pymnt_inv, total_pymnt - They have no impact on loan status to be charged off. so are removed.

total_rec_prncp, total_rec_int, total_rec_late_fee, collection_recovery_fee, recoveries, last_pymnt_d, last_pymnt_amnt, last_credit_pull_d, out_prncp are all calculated after the loan being charged off. So they are removed.

```python
#Remove the unnecessary columns and check null percentage
loan_colfilter_df = loan_drop_null_df.drop(['next_pymnt_d','mths_since_last_record','member_id','funded_amnt','funded_amnt_inv','url','desc','zip_code','mths_since_last_delinq','revol_bal','out_prncp_inv','total_pymnt_inv','total_pymnt','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d','out_prncp'],axis=1)
round(100*(loan_colfilter_df.isnull().sum()/len(loan_colfilter_df.index)), 2)
```

```python
#Remove the starting and leading whitespaces for string objects
str_cols_df = loan_colfilter_df.select_dtypes(['object'])
loan_colfilter_df[str_cols_df.columns] = str_cols_df.apply(lambda x: x.str.strip())
loan_colfilter_df.head()
```

```python
#Convert all string columns data to uppercase
loan_colfilter_df[str_cols_df.columns] = str_cols_df.apply(lambda x: x.str.upper())
loan_colfilter_df.head()
```

```python
#Describe the observations in numeric columns
print(loan_colfilter_df.describe())
```

It can be observed that tax_liens, chargeoff_within_12_mths, collections_12_mths_ex_med, policy_code, acc_now_delinq, delinq_amnt all represent a singular value in all the rows. SO they can be dropped.

```python
#Drop the columns where there is no change in data and check the null percentage
loan_colfilter_df = loan_colfilter_df.drop(['tax_liens','chargeoff_within_12_mths','collections_12_mths_ex_med','policy_code','acc_now_delinq','delinq_amnt'],axis=1)
round(100*(loan_colfilter_df.isnull().sum()/len(loan_colfilter_df.index)), 2)
```

```python
#Check the column details post columns drop
loan_colfilter_df.info()
```

Get the details of the 'Object' type columns to check if there is any data which is of no significance

```python
print(loan_colfilter_df['term'].describe())
print(loan_colfilter_df['int_rate'].describe())
print(loan_colfilter_df['grade'].describe())
print(loan_colfilter_df['sub_grade'].describe())
print(loan_colfilter_df['emp_title'].describe())
print(loan_colfilter_df['emp_length'].describe())
print(loan_colfilter_df['home_ownership'].describe())
print(loan_colfilter_df['verification_status'].describe())
print(loan_colfilter_df['issue_d'].describe())
print(loan_colfilter_df['loan_status'].describe())
print(loan_colfilter_df['pymnt_plan'].describe())
print(loan_colfilter_df['purpose'].describe())
print(loan_colfilter_df['title'].describe())
print(loan_colfilter_df['addr_state'].describe())
print(loan_colfilter_df['earliest_cr_line'].describe())
print(loan_colfilter_df['revol_util'].describe())
print(loan_colfilter_df['initial_list_status'].describe())
print(loan_colfilter_df['application_type'].describe())
```

emp_title, title - Having too many values and its very difficult to categorize. Can be removed.
pymnt_plan, application_type, initial_list_status - They all represent singular value. Can be removed.

```python
#Drop columns which are not having significance or singular data
loan_colfilter_df = loan_colfilter_df.drop(['emp_title','pymnt_plan','title','application_type','initial_list_status'],axis=1)
loan_colfilter_df.info()
```

```python
#Check null percentage
round(100*(loan_colfilter_df.isnull().sum()/len(loan_colfilter_df.index)), 2)
```

```python
#Get the number of null rows in revol_util,emp_length,pub_rec_bankruptcies   columns, remove them if they are small and check null percentage
print(loan_colfilter_df['revol_util'].isnull().sum())
print(loan_colfilter_df['emp_length'].isnull().sum())
print(loan_colfilter_df['pub_rec_bankruptcies'].isnull().sum())

loan_colfilter_df=loan_colfilter_df[~(pd.isnull(loan_colfilter_df['revol_util']))]
loan_colfilter_df=loan_colfilter_df[~(pd.isnull(loan_colfilter_df['emp_length']))]
loan_colfilter_df=loan_colfilter_df[~(pd.isnull(loan_colfilter_df['pub_rec_bankruptcies']))]

round(100*(loan_colfilter_df.isnull().sum()/len(loan_colfilter_df.index)), 2)
```

Since int_rate and revol_util represent percentages remove the percentage symbol and make them numeric

```python
#Split the values by % to new columns and convert to float type
loan_colfilter_df['int_rate_num']=loan_colfilter_df['int_rate'].apply(lambda x : x.split('%')[0])
loan_colfilter_df['int_rate_num']=loan_colfilter_df.int_rate_num.astype('float64')
loan_colfilter_df['revol_util_num']=loan_colfilter_df['revol_util'].apply(lambda x : x.split('%')[0])
loan_colfilter_df['revol_util_num']=loan_colfilter_df.revol_util_num.astype('float64')
loan_colfilter_df.head()
```

```python
#Get the months value from term to a new column and convert the type to integer
loan_colfilter_df['term_months']=loan_colfilter_df['term'].apply(lambda x: x[:2])
loan_colfilter_df['term_months'] = loan_colfilter_df['term_months'].astype('int64')
loan_colfilter_df['term_months'].describe()
```

```python
#Get the year and months value from issue_d to new columns
loan_colfilter_df['issue_d_month'], loan_colfilter_df['issue_d_yr'] = loan_colfilter_df['issue_d'].str.split('-', 1).str
loan_colfilter_df.head()
```

```python
#Get the year and months value from earliest_cr_line to new columns
loan_colfilter_df['earliest_cr_line_month'], loan_colfilter_df['earliest_cr_line_yr'] = loan_colfilter_df['earliest_cr_line'].str.split('-', 1).str
loan_colfilter_df.head()
```

```python
#Drop the original columns to a new dataframe
loan_trim_df = loan_colfilter_df.drop(['term','int_rate','revol_util','issue_d','earliest_cr_line'],axis=1)

#Get the frequency of values of loan_status
loan_trim_df.loan_status.value_counts()
```

```python
#We are not finding potential bad loans. So lans with 'Current' status can be ignored.
loan_trim_df=loan_trim_df[loan_trim_df.loan_status!='CURRENT']
loan_trim_df.loan_status.value_counts()
```

```python
#Check the info of the columns after handling the data
loan_trim_df.info()
```

### Analysis on Categorical variables

Get the number and percentage of fully paid and charged off loans for each column to understand their effect on the loan status.

##### Grade Column

```python
#Get the value counts of each grade
print(loan_trim_df.grade.value_counts())

#Group the grades by loan status
grade_groupby_df= loan_trim_df.groupby(['grade','loan_status']).count()

#Get the percentage of fully paid vs charged off across the grades
grade_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id
```

From the plot it can be observed that the charge off rate increases as the grade increases.
So 'A' represents the least charge off value grade while 'G' represents the highest charge off grade.

```python

# set seaborn theme
sns.set(style="white")
#Setting figure size and initializing subplots
plt.subplots(figsize=(20,5))
plt.subplot(1, 2, 1)

#Count plot with grades against loans
axis_count = sns.countplot(x='grade',hue='loan_status' , data=loan_trim_df, order=loan_trim_df['grade'].value_counts().index)
axis_count.set(xlabel='Grade', ylabel='Count of loans')
plt.title('Count of Loan Status against Grades') 

#Get percentage of charged off vs fully paid loans for each grade   
grade_df = grade_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id.reset_index()

#initializing subplot
plt.subplot(1, 2, 2)

#Bar plot with grades vs percent of charged off loans
axis_count = sns.barplot(x='grade',y='id',hue='loan_status' , data=grade_df[grade_df.loan_status=='CHARGED OFF'],palette=["C1"])
axis_count.set(xlabel='Grade', ylabel='percentage of Charged Off loans')
plt.title('% of Charged Off Loans against Grades') 
plt.show()
```

##### sub_grade column

```python
#Get the value counts of each sub grade
print(loan_trim_df.sub_grade.value_counts())

#Group the sub grade by loan status for each sub grade
subgrade_groupby_df= loan_trim_df.groupby(['sub_grade','loan_status']).count()

#Get the percentage of fully paid vs charged off across the sub grades
subgrade_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id
```

From the plot the trend of increase in charge of rate with sub grades can be observed


```python
#Setting figure size and initializing subplots
plt.subplots(figsize=(20,5))
plt.subplot(2, 1, 1)

#Count plot with sub grades against loans
axis_count = sns.countplot(x='sub_grade',hue='loan_status' , data=loan_trim_df, order=loan_trim_df['sub_grade'].value_counts().index)
axis_count.set(xlabel='Sub Grades', ylabel='Count of loans')
plt.title('Count of Loan status against sub grades') 

#Get percentage of charged off vs fully paid loans for each sub grade 
subgrade_df = subgrade_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id.reset_index()

#Initialize subplot
plt.subplot(2, 1, 2)

#Bar plot with sub grades vs percent of charged off loans
axis_count = sns.barplot(x='sub_grade',y='id',hue='loan_status' , data=subgrade_df[subgrade_df['loan_status']=='CHARGED OFF'],palette=["C1"])
axis_count.set(xlabel='Sub Grades', ylabel='Percentage of Charged Off Loans')
plt.title('% of Charged Off loans against sub grades') 
plt.show()
```

Now merging grade and sub grade and checking the charge off percentage can reveal the relation between grades and sub grades and their effect on the charged off loan status

```python
#Group the dataframe by grades and sub grades
grade_subgrade_group= loan_trim_df.groupby(['grade','sub_grade','loan_status']).count()

#Get the percentage of loan status against the grade and sub_grade group
grade_subgrade_group.groupby(level=[0,1]).apply(lambda x: 100*(x/x.sum())).id
```

It can be observed from the table above that charge off rate increases within each subgrade against each grade. The same be observed from the plot below.

```python
#Get the percentage of loan status against the grade and sub_grade group
grade_subgrade_df = grade_subgrade_group.groupby(level=[0,1]).apply(lambda x: 100*(x/x.sum())).id.reset_index()

#Initialize the figure size and sub plot
plt.subplots(figsize=(20,5))
plt.subplot(1, 1, 1)

#Bar plot with percentage of charged off loans against combination of grades and sub_grades
axis_count = sns.barplot(x='grade',y='id',hue='sub_grade' , data=grade_subgrade_df[grade_subgrade_df['loan_status']=='CHARGED OFF'])
axis_count.set(xlabel='Sub Grades', ylabel='% of Charged Off Loans')
plt.title('Grade & Sub Grades vs of Charged Off loans') 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

##### home_ownership analysis

```python
#Get value counts for each home ownership
print(loan_trim_df.home_ownership.value_counts())

#Group the dataframe by home_ownership and loan_status
home_own_groupby_df= loan_trim_df.groupby(['home_ownership','loan_status']).count()

#Get the percentage of loan status against the home_ownership
home_own_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id
```

The table reveals some interesting insights.

Mortgage -  This has the least charge off rate - Probably the loans are paid off with the bank taking the possession of house ownership. This explains why the charge off rate is least for mortgaged houses. 

Own - This has the least charge off rate excluding mortgage. Moreover loans against owned property can be mortgaged to pay off the debt so this represents the least risk for charged off loans

Rent and Other - These represent the highest charge off rates as expected as the debt to income ratio increases with the rental payments.

Also mostof the loans are given to the people with rental properties thus contributing to higher charge off rates.

The same can be visualized from the plots below

```python
#Initialize the figure size and sub plot
plt.subplots(figsize=(20,5))
plt.subplot(2, 1, 1)

#Count plot with sub home_ownership against loan status
axis_count = sns.countplot(x='home_ownership',hue='loan_status' , data=loan_trim_df, order=loan_trim_df['home_ownership'].value_counts().index)
axis_count.set(xlabel='Home Ownership', ylabel='Count of loans')
plt.title('No of loans against Home Ownership') 

#Plot in log scale as there is huge variation in values
plt.yscale('log')

#Get the percentage of loan status against the home_ownership
verification_df = home_own_groupby_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id.reset_index()

#Initialize the sub plot
plt.subplot(2, 1, 2)

#Bar plot with percentage of charged off loans against home_ownership
axis_count = sns.barplot(x='home_ownership',y='id',hue='loan_status' , data=verification_df[verification_df['loan_status']=='CHARGED OFF'],palette=["C1"])
axis_count.set(xlabel='Home Ownership', ylabel='Percentage of loans')
plt.title('% of Charged off loans against home ownership') 
plt.yscale('log')

plt.show()
```

##### earliest_cr_line column analysis

Since there wont be any impact on loan status on which month the credit line analysis is done, we will only evaluate the year when the credit analysis was done

```python
#Get no of loans issued against each year credit line was analysed
print(loan_trim_df.earliest_cr_line_yr.value_counts())

#Group the credit line verified years and loan status
earliest_cr_line_yr_df= loan_trim_df.groupby(['earliest_cr_line_yr','loan_status']).count()

#Get percentage of loan status against earliest credit line year
earliest_cr_line_yr_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id
```

There are many values with very few records and they might skew the data. So lets consider the values where the no of loans issued are greater than 100

```python
#Filter the years where the number of issued loans are greater than 100 and get the data into a new dataframe
cr_line_filter = loan_trim_df.earliest_cr_line_yr.value_counts()>100
cr_line_filter_df = pd.DataFrame(cr_line_filter).reset_index()
cr_line_filter_df=cr_line_filter_df[cr_line_filter_df['earliest_cr_line_yr']==True]
cr_line_filter_df['earliest_cr_line_yr']=cr_line_filter_df['index'].get_values()
cr_line_filter_df = loan_trim_df[loan_trim_df.earliest_cr_line_yr.isin(cr_line_filter_df.earliest_cr_line_yr)]

#Group the credit line verified years and loan status
cr_line_df = cr_line_filter_df.groupby(['earliest_cr_line_yr','loan_status']).count()

#Get percentage of loan status against earliest credit line year
print(cr_line_df.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).id)
```

Lets plot the results and see if we can find any pattern to analyze

##### installment and annual_inc columns analysis

lets create a new derived column called installment_percent which tells what percent of annual income the users are paying as installment. This could be a good measure to identify the percentage of charged off loans.

Also the new derived column is proportional to installment and annual_inc we dont need to do seperate column analysis on each of these two columns

```python
#Create a new column installment_percent as a ratio of installment to annual income in percentage
loan_trim_df['installment_percent'] = round((100*loan_trim_df['installment']*12)/loan_trim_df['annual_inc'],2)
print(loan_trim_df.head())

loan_trim_df['installment_percent'].describe()
```

Since the values ranges from 0 to 30 let's divide the values in groups of ranges in increments of 5

As expected as installment percent of their annual income increases the charge off rates increase. If the installment ratio higher than 5 poses significant risk in charged off loans. The slight dip in the ratio range after 20 may be because of less number of total loans processed in that range.

```python
#Initialize figure and subplots
plt.subplots(figsize=(20,5))
plt.subplot(1, 1, 1)

#Get % of loan status aganst the group of loan amount and interest rate groups
loan_int_group_df = loan_int_group.groupby(level=[0,1]).apply(lambda x: 100*(x/x.sum())).reset_index()

#Bar plot the percentage of charged off loans against the group of loan amount and interest rate groups
axis_count = sns.barplot(x='loan_group',y='id',hue='int_rate_num_group' , data=loan_int_group_df[loan_int_group_df.loan_status=='CHARGED OFF'])
axis_count.set(xlabel='Loan Group', ylabel='percentage of loans')

plt.title('% of Charged Off loans against groups of loans and interest rate') 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

This gives a great insight on the impact of interest rates agains the charged off loans. Regardless of loan amounts the higher the interest rates the higher the rate of charged off loans. So interest rate is a significant charged off loans identifier metric than that of loan amount

##### revol_util_num column analysis

```python
#Get the count of values of revol_util_num
print(loan_trim_df.groupby(['revol_util_num','loan_status']).id.count())

#Get observations on revol_util_num
loan_trim_df.revol_util_num.describe()
```

Since the values range between o to 99 lets bin the values in the range of 0 to 105 in increments of 10

```python
#Bin the values in increments of 10 in the range of 0 to 105
loan_trim_df['revol_util_num_group'] = pd.cut(loan_trim_df['revol_util_num'], range(0, 105, 10), right=False, include_lowest=True)

#Group the binned revolving util values to loan status
revol_util_num_group=loan_trim_df.groupby(['revol_util_num_group','loan_status']).id.count()

#Get % of loan status against revolving utilization value
revol_util_num_group.groupby(level=[0]).apply(lambda x: 100*(x/x.sum()))
```

This shows as the credit card utilization increases the charged off loans too increase. We can see below that the percentage of non-performing loans steadily increases with utilization rate. Borrowers with high utilization rates are more likely to have high fixed credit card payments which might affect their ability to repay their loans. Also, a high utilization rate often reflects a lack of other financing options, with borrowers turning to peer-to-peer lending as a last resort. This is in contrast to those borrowers with low utilization rates, who may be using peer-to-peer lending opportunistically to pursue lower interest payments. 

Let's see the analysis in a plot

```python
#Initialize figure and subplots
plt.subplots(figsize=(20,5))
plt.subplot(1, 2, 1)

#Count plot with revol_util_num groups against loan status
axis_count = sns.countplot(x='revol_util_num_group',hue='loan_status' , data=loan_trim_df, order=loan_trim_df['revol_util_num_group'].value_counts().index)
axis_count.set(xlabel='Revolving line util', ylabel='Count of loans')
plt.title('No of loans against credit card revolving utilization') 

#Get % of loan status against revolving utilization value
revol_util_num_group_df = revol_util_num_group.groupby(level=[0]).apply(lambda x: 100*(x/x.sum())).reset_index()

#Initialize sub plot
plt.subplot(1, 2, 2)
axis_count = sns.barplot(x='revol_util_num_group',y='id',hue='loan_status' , data=revol_util_num_group_df[revol_util_num_group_df.loan_status=='CHARGED OFF'],palette=["C1"])
axis_count.set(xlabel='Revolving line util', ylabel='percentage of charged off loans')
plt.title('% of charged off loans against credit revolving utilization') 
plt.show()
```
